```
# Import library
import numpy as np
import scipy.io as scio
from sklearn.svm import SVC
import matplotlib as mpl
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
%matplotlib inline
```

```
# Dataset1
data1 = scio.loadmat('ex6data1.mat')
X1 = data1['X']
y1 = data1['y'].flatten()
m = y1.size
print ("X1 shape: " + str(X1.shape))
print ("y1 shape: " + str(y1.shape))
```

```
X1
```

```
y1
```

```
# Visualize data
plt.scatter(X1[:, 0], X1[:, 1]);
```

```
# Linear SVM
svm = SVC(kernel='linear', C=100, random_state=1)
svm.fit(X1, y1);
```

```
# Plotting decision regions
plot_decision_regions(X1, y1, clf=svm)
plt.show()
```

```
# Implement gaussian kernel from scratch
def gaussian_kernel(x1, x2, sigma=0.1):
    # Ensure x1 and x2 are column vectors
    x1 = x1.flatten()
    x2 = x2.flatten()

    # Initialize x1 and x2 similarity
    similarity = 0
    similarity = np.exp(-np.sum(np.square(x1-x2))/(2*np.square(sigma)))
  
    return similarity
```

```
# Test gaussian_kernel()
x1 = np.array([1, 2, 1])
x2 = np.array([0, 4, -1])
similarity = gaussian_kernel(x1, x2, sigma=2)
print("similarity between x1 and x2: " +str(similarity))
print("This value should be about 0.324652")
```

```
# Dataset2
data2 = scio.loadmat('ex6data2.mat')
X2 = data2['X']
y2 = data2['y'].flatten()
m = y2.size
print ("X2 shape: " + str(X2.shape))
print ("y2 shape: " + str(y2.shape))
```

```
# Visualize data
plt.scatter(X2[:, 0], X2[:, 1]);
```

```
# SVM with gaussian kernels
svm = SVC(kernel='rbf', C=100, random_state=1, gamma=0.10) # gamma -- gaussian kernels terms /(2*np.square(sigma))
svm.fit(X2, y2);
```

```
# Plotting decision regions
plot_decision_regions(X2, y2, clf=svm)
plt.show();
```
